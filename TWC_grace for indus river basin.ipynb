```python
# Calculate the long-term mean
jpl_mean = df['Water Equivalent Thickness – Land (GRACE, GRACE-FO JPL)'].mean()
csr_mean = df['Water Equivalent Thickness – Land (GRACE, GRACE-FO CSR)'].mean()

# Calculate anomalies
df['JPL Anomaly'] = df['Water Equivalent Thickness – Land (GRACE, GRACE-FO JPL)'] - jpl_mean
df['CSR Anomaly'] = df['Water Equivalent Thickness – Land (GRACE, GRACE-FO CSR)'] - csr_mean

# Display the DataFrame with anomalies
print(df.head())
```

```python
import numpy as np
from scipy.stats import linregress

# Calculate linear trend for JPL data
slope_jpl, intercept_jpl, r_value_jpl, p_value_jpl, std_err_jpl = linregress(np.arange(len(df)), df['Water Equivalent Thickness – Land (GRACE, GRACE-FO JPL)'])

# Calculate linear trend for CSR data
slope_csr, intercept_csr, r_value_csr, p_value_csr, std_err_csr = linregress(np.arange(len(df)), df['Water Equivalent Thickness – Land (GRACE, GRACE-FO CSR)'])

print(f"JPL Trend: Slope = {slope_jpl}, Intercept = {intercept_jpl}, R-value = {r_value_jpl}, P-value = {p_value_jpl}")
print(f"CSR Trend: Slope = {slope_csr}, Intercept = {intercept_csr}, R-value = {r_value_csr}, P-value = {p_value_csr}")
```

```python
from pymannkendall import original_test

# Perform Mann-Kendall test for JPL data
mk_test_jpl = original_test(df['Water Equivalent Thickness – Land (GRACE, GRACE-FO JPL)'])
print(f"Mann-Kendall Test for JPL Data: {mk_test_jpl}")

# Perform Mann-Kendall test for CSR data
mk_test_csr = original_test(df['Water Equivalent Thickness – Land (GRACE, GRACE-FO CSR)'])
print(f"Mann-Kendall Test for CSR Data: {mk_test_csr}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot the data and anomalies
plt.figure(figsize=(14, 8))

# Plot JPL data and anomalies
plt.subplot(2, 1, 1)
plt.plot(df['Date'], df['Water Equivalent Thickness – Land (GRACE, GRACE-FO JPL)'], label='JPL Data')
plt.plot(df['Date'], df['JPL Anomaly'], label='JPL Anomaly', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Water Equivalent Thickness')
plt.title('JPL Data and Anomalies')
plt.legend()

# Plot CSR data and anomalies
plt.subplot(2, 1, 2)
plt.plot(df['Date'], df['Water Equivalent Thickness – Land (GRACE, GRACE-FO CSR)'], label='CSR Data')
plt.plot(df['Date'], df['CSR Anomaly'], label='CSR Anomaly', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Water Equivalent Thickness')
plt.title('CSR Data and Anomalies')
plt.legend()

plt.tight_layout()
plt.show()

# Plot trends
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Water Equivalent Thickness – Land (GRACE, GRACE-FO JPL)'], label='JPL Data')
plt.plot(df['Date'], intercept_jpl + slope_jpl * np.arange(len(df)), label='JPL Trend', linestyle='--')
plt.plot(df['Date'], df['Water Equivalent Thickness – Land (GRACE, GRACE-FO CSR)'], label='CSR Data')
plt.plot(df['Date'], intercept_csr + slope_csr * np.arange(len(df)), label='CSR Trend', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Water Equivalent Thickness')
plt.title('Trend Analysis')
plt.legend()
plt.show()
```
