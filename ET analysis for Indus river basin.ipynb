```python
# Group by month and calculate mean ET
monthly_et = df_long.groupby(["Month", "Year"])["ET"].mean().reset_index()

# Display monthly ET data
monthly_et.head()
```

```python
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Initialize a dictionary to store regression results
monthly_trends = {}

# Plot monthly trends
plt.figure(figsize=(14, 10))
for i, month in enumerate(monthly_et["Month"].unique(), 1):
    # Filter data for the current month
    month_data = monthly_et[monthly_et["Month"] == month]
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(month_data["Year"], month_data["ET"])
    
    # Store results
    monthly_trends[month] = {
        "Slope": slope,
        "Intercept": intercept,
        "R-squared": r_value**2,
        "P-value": p_value
    }
    
    # Plot the data and trend line
    plt.subplot(4, 3, i)
    plt.plot(month_data["Year"], month_data["ET"], marker="o", linestyle="-", color="blue", label=f"{month} ET")
    plt.plot(month_data["Year"], intercept + slope * month_data["Year"], linestyle="--", color="red", label="Trend Line")
    plt.title(f"{month} Evapotranspiration Trends")
    plt.xlabel("Year")
    plt.ylabel("ET (mm)")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Display regression results for each month
for month, results in monthly_trends.items():
    print(f"Month: {month}")
    print(f"  Slope (Trend): {results['Slope']:.4f}")
    print(f"  P-value: {results['P-value']:.4f}")
    print(f"  R-squared: {results['R-squared']:.4f}")
    print()
```
